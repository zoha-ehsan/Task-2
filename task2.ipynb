```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("titanic_dataset.csv")
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df["Fare"].fillna(df["Fare"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()    #summary statistics
```

```python
df.groupby("Sex")["Survived"].sum()  #survival count by gender
```

```python
df.groupby("Pclass")["Survived"].sum()    #survival count by class
```

```python
survival_rate = df.groupby("Sex")["Survived"].mean()
survival_count = df.groupby("Sex")["Survived"].sum()
survival_table = df.groupby(["Sex", "Survived"]).size().unstack()
survival_table.plot(kind="bar", stacked=True, color=["grey", "black"])
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(["Did not Survive", "Survived"])
plt.show()                                                                               #Survival rate by gender
```

```python
sns.barplot(data=df, x="Pclass", y="Survived", ci=None, palette="muted")
plt.title("Survival rate by passenger class")
plt.show()
```

```python
sns.barplot(data=df, x="Embarked", y="Survived", ci=None, palette="Set2")
plt.title("Survival Rate by Port of Embarkation")
plt.show()
```

```python
corr = df[["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(survival_counts,
        labels=['Not Survived', 'Survived'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['grey', 'pink'])
plt.title("Survival Distribution")
plt.show()
```

```python
numeric_cols = ["Age", "Fare"]
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df[df["Survived"]==1]["Age"], fill=True, color="green", label="Survived")
sns.kdeplot(data=df[df["Survived"]==0]["Age"], fill=True, color="red", label="Not Survived")
plt.title("Age Distribution by Survival Status")
plt.legend()
plt.show()
```

```python
sns.pairplot(df[["Survived", "Pclass", "Age", "Fare"]], hue="Survived", palette="coolwarm")
plt.show()
```
